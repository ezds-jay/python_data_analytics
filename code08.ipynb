````python
# 데이터 읽어오기
df = pd.read_csv("data/ibm-hr-attrition.csv")

# head : 데이터의 상위 row를 보여줍니다.
df.head()
````

````python
# 전체 데이터의 크기를 볼 수 있습니다.

df.shape
````

````python
# 인덱싱
# 인덱싱이란 원하는 정보들을 보는 방법을 얘기합니다.

# 0번행의 Age 정보
df.loc[0,'Age']
````

````python
# iloc을 이용해서 인덱스를 활용해서 필요한 데이터 부분을 살펴볼 수 있다.

df.iloc[1:5,2:5]
````

````python
# 선택한 컬럼들만 보고 싶을때, 컬럼의 이름을 이용합니다.

df[['BusinessTravel', 'DailyRate', 'Education']]
````

````python
# 만일 새로 편집한 데이터를 index를 새로하고 싶다면
# reset_index 방법을 씁니다.

df.loc[10:20, ["BusinessTravel", "DailyRate"]].reset_index(drop=True)
````

````python
# 새로운 컬럼을 추가하려면
# 생성된 인스턴스에 [변수이름]을 포함한, 전체 행 개수와 동일한 길이를 가진 리스트를 넣어주면 됩니다.
import random

random.seed(1)
nrows = len(df)
val = [random.choice(["apple", "orange", "kiwi"]) for i in range(0, nrows)]

df["fruits"] = val

# 아래에 새로 컬럼이 추가된 것을 확인할 수 있다. 
df.iloc[:10, -5:]
````

````python
# 만일 컬럼 이름을 변경하고 싶다면
# rename이라는 메서드를 이용합니다.

df = df.rename(columns = {"fruits" : "과일"})
df.iloc[:10, -5:]
````

````python
# 컬럼의 데이터를 새롭게 수정하고 싶을 때
# apply method를 사용할 수 있습니다.
# 과일 컬럼에서 다음과 같이 수정한다고 해봅시다.
# apple -> strawberry
# orange -> blueberry
# kiwi -> cranberry

def convert(x) :
    if x == 'apple' :
        return 'strawberry'
    elif x == 'orange' :
        return 'blueberry'
    else :
        return 'cranberry'

df['과일'].apply(convert)
````

각 컬럼별로 집계하는 방법
- 숫자형(연속형)
    - 평균, 중간값, 분산(편차), 최대최소, 사분위 범위
    - 빈도(분포)

- 이산형(카테고리형)
    - 빈도수, 퍼센티지(%)
    - 막대그래프

````python
# mean -> 평균

df["Age"].mean()
````

````python
# median -> 중간값
df['Age'].median()
````

````python
# 표준편차(분산)

df['Age'].var()
````

````python
# 최대/최소

df["Age"].max(), df["Age"].min()
````

````python
# 분포도

df["Age"].hist(bins=30)
````

````python
# 빈도수 

df["Attrition"].value_counts()
````

````python
# 비율 

cnt = df["Attrition"].count()
df["Attrition"].value_counts() / cnt
````

데이터를 필터링하는 방법

데이터 집계 방법 : groupby

````python
# Department에 따라서 age의 분포를 보고 싶다면
# groupby를 사용해볼 수 있습니다.

df.groupby(['Department'], as_index=False)['Age'].mean()
````

````python
# 평균, 중간값, 분산을 다같이 보고 싶다면

df.groupby(['Department'], as_index=False)['Age'].agg(['mean', 'median', 'var'])
````
